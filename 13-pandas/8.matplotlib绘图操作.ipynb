## 柱状图：plt.bar()
- 参数：1.索引  2.数据值  3.条形宽度

## 直方图  （是一个特殊的柱状图，又叫密度图）
- plt.hist()参数:
    - bins :可以是一个bin数量的整数值，也可以是标识bin的一个序列。默认值=10
    - normed :如果值=True，直方图的值将进行归一化处理，形成概率密度，默认值=False
    - color: 颜色，可以是单一颜色值或颜色的序列，如果指定了多个数据集合，例如DataFrame对象，颜色序列将会设置为相同的顺序。如果为未指定，将会使用一个默认的颜色线条
    - orientation: 通过设置orientation=horizontal创建水平直方图，默认值=vertical

## 饼图 pie() 适合展示各部分占总体的比例，条形图适合比较各部分的大小
- labeldistance label到圆心距离  autopct 所占比例  explode 每个组成部分到圆心距离

## 散点图 scatter()
- 因变量随自变量而变化的大致趋势


```python
import matplotlib.pyplot as plt
import numpy as np
```

#### 绘制单条线形图

```python
x = np.array([1,2,3,4,5])
y = x+3
plt.plot(x,y)
```

#### 绘制多条线形图

```python
plt.plot(x,y)
plt.plot(x+1,y-1)
```

#### 设置坐标系的比例plt.figure(figsize=(a,b))

```python
plt.figure(figsize=(5,9)) #放在plot前面
plt.plot(x,y)
```

#### 设置图例legend()

```python
plt.plot(x,y,label='x,y')
plt.plot(x+1,y-1,label='x+1,y-1')
plt.legend()
```

#### 设置轴的标识

```python
plt.plot(x,y,label='x,y')
plt.xlabel('temp')
plt.ylabel('dist')
plt.title('title')
```

#### 图例保存

```python
fig = plt.figure()
plt.plot(x,y)
fig.savefig('../123456.png')
```

```python
# 曲线的样式和风格（自学）
plt.plot(x,y,c='red',alpha=0.5,linestyle='dashed')
```

### 柱状图：plt.bar()

```python
plt.bar(x,y)
# plt.barh(x,y) #横向
```

## 直方图 plt.hist() （是一个特殊的柱状图，又叫密度图）

```python
data = [1,1,2,2,3,4,5,6,7,7,7,7,8,8,6,5,0,9]
plt.hist(data, bins=20)
```

#### 饼图 pie()

```python
arr = [15,18,20,30]
plt.pie(arr,labels=['a','b','c','d'],labeldistance=0.5,autopct='%.2f%%')
```

#### 散点图 scatter()
- 因变量随自变量而变化

```python
x = np.array([1,3,5,7,9])
y = x**2 -3
plt.scatter(x,y)
```
